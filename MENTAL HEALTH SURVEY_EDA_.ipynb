## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Filter for realistic age values (already cleaned)
df = df[(df['Age'] > 16) & (df['Age'] < 100)]


# Clean and standardize gender entries
# Step 1: Reload fresh original data
df = pd.read_csv('/content/drive/MyDrive/Mental Health Survey EDA Analysis/survey.csv')
# Step 2: Lowercase and strip
df['Gender'] = df['Gender'].str.lower().str.strip()

# Step 3: Preview unique values BEFORE cleaning
print(df['Gender'].unique())  #  This time you'll see: ['male', 'female', 'f', 'woman', ...]

# Step 4: Clean correctly
def clean_gender(g):
    if any(word in g for word in ['female', 'f', 'woman']):
        return 'female'
    elif any(word in g for word in ['male', 'm', 'man']):
        return 'male'
    elif 'trans' in g:
        return 'trans'
    else:
        return 'other'

df['Gender'] = df['Gender'].apply(clean_gender)

# Step 5: Check the final distribution
print(df['Gender'].value_counts())
df['Age'].describe()
````

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code

# Create a copy of the DataFrame to avoid modifying the original during encoding
df_corr = df.copy()

# Encode 'treatment' (and potentially other relevant categorical features) to numerical for correlation calculation
df_corr['treatment_encoded'] = df_corr['treatment'].apply(lambda x: 1 if x == 'Yes' else 0)

# Select numerical columns for correlation, including the encoded 'treatment'
numerical_cols = ['Age', 'treatment_encoded']

# Identify other potentially numerical or ordinal categorical columns that could be encoded
# For this dataset, 'no_employees', 'leave' might have an ordinal sense.
# Let's consider 'no_employees' as it has ordered categories.

# Mapping for 'no_employees' for ordinal encoding (example mapping, adjust as needed)
no_employees_mapping = {
    '1-5': 1,
    '6-25': 2,
    '26-100': 3,
    '100-500': 4,
    '500-1000': 5,
    'More than 1000': 6
}
df_corr['no_employees_encoded'] = df_corr['no_employees'].map(no_employees_mapping)
numerical_cols.append('no_employees_encoded')

# Calculate the correlation matrix
correlation_matrix = df_corr[numerical_cols].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features and Treatment')
plt.show()
````

##### 1. Why did you pick the specific chart?

A correlation heatmap was chosen because it provides a clear, visual representation of the correlation matrix between multiple numerical variables. It allows for quick identification of strong relationships (positive or negative) between variables, which is crucial for understanding the dataset's structure and potential dependencies. This helps in spotting variables that might influence the 'treatment' outcome.

##### 2. What is/are the insight(s) found from the chart?

From the correlation heatmap, we can observe:
*   **Age and Treatment:** There is a very weak positive correlation (approximately 0.07) between 'Age' and 'treatment_encoded'. This suggests that age has a negligible linear impact on whether an individual seeks mental health treatment.
*   **Number of Employees and Treatment:** The correlation between 'no_employees_encoded' and 'treatment_encoded' is extremely low (approximately 0.01). This indicates virtually no linear relationship between the size of the employer's company and the likelihood of seeking treatment.
*   **Age and Number of Employees:** There is a weak positive correlation (approximately 0.12) between 'Age' and 'no_employees_encoded'. While this is the strongest correlation among the numerical variables shown, it still suggests a minor linear relationship, meaning that older respondents might tend to work for slightly larger companies, or vice versa.

A pair plot is chosen because it allows for a quick visualization of the pairwise relationships between multiple variables in the dataset. It displays scatter plots for each pair of variables and histograms for each single variable, providing insights into their distributions and correlations at a glance.

##### 2. What is/are the insight(s) found from the chart?

From the pair plot, we can observe:
*   **Distributions:** The histograms on the diagonal show the distribution of each variable. For instance, we can see the age distribution and how the number of employees is distributed across the defined categories.
*   **Pairwise Relationships:** The scatter plots show how each pair of variables is related. For example, we can observe if there's any visible pattern between 'Age' and 'no_employees_encoded', or how 'treatment_encoded' might separate or group these variables.
*   **Influence of Treatment:** By coloring the points based on 'treatment_encoded' (whether treatment was sought or not), we can visually inspect if respondents who sought treatment ('Yes') tend to cluster in specific age groups or company sizes compared to those who did not ('No').
